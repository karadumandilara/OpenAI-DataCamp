```python
## Sentiment

prompt = """
Classify sentiment in the following statements and after that classify sentiment as 1 - 5(bad-good):
1. Servis çok yavaştı
2. Et baya lezzetliydi
3. Yemek fena değildi ama daha iyisini yemiştim.
4. Yemeğim gecikti ama içecekler iyiydi.
"""

response = openai.Completion.create(
  model="gpt-3.5-turbo-instruct",
  prompt= prompt,
  max_tokens = 50,
  temperature = 1 ## gerek olmayabilir

)
print(response.choices[0]["text"])
```

```python
## Zero-shot vs. one-shot vs. few-shot prompting (in context learning)

### one shot: 
prompt = """
Classify sentiment in the following statements:
1. Servis çok yavaştı // Pozitif
2. Yemek fena değildi ama daha iyisini yemiştim. //
"""
response = openai.Completion.create(
  model="gpt-3.5-turbo-instruct",
  prompt= prompt
)
print(response.choices[0]["text"])
```

```python
### few shot: 
prompt = """
Classify sentiment in the following statements:
1. Servis çok yavaştı // Negatif
2. Yemek fena değildi ama daha iyisini yemiştim. // Nötr
3. Et baya lezzetliydi // Pozitif
4. Yemeğim gecikti ama içecekler iyiydi. //

"""
response = openai.Completion.create(
  model="gpt-3.5-turbo-instruct",
  prompt= prompt
)
print(response.choices[0]["text"])
```

#### roles
- system: controls asistant's *behavior*
- user: *instruct* the assistant
- assistant: response to user instruction
    - can also be written by the user to provide examples

```python
response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [{"role":"system",
    "content":"you are a fashion designer at miu miu"},
    {
        "role":"user",
        "content":"Can you suggest me good fashion brands that use of high quality fabrics? "
    }]
)
```

```python
## Multi-turn chat completions with GPT
```

```python
response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [{"role":"system",
    "content":"you are a fashion designer at miu miu"},

    {
        "role":"user",
        "content":"I need little bit cheaper brands than high fashion brands? "
    },
    {
        "role":"assistant",
        "content":"such as GUESS"
    },

    {
        "role":"user",
        "content":"Can you suggest me good fashion brands that use of high quality fabrics? "
    }]
)
```

```python
response["choices"][0]["message"]["content"]
```

```python
## Storing responses

messages = [{"role":"system","content":"You are a data science tutor who provides short, simple explanation."}]
user_qs = ["Why is Python so popular?", "Summarize this in one sentence."]

for q in user_qs:
    print("User: ", q)
    user_dict= {"role":"user","content":q}
    messages.append(user_dict)
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = messages
    )
    assistant_dict = dict(response["choices"][0]["message"])
    messages.append(assistant_dict)
    print("Assistant: ", response["choices"][0]["message"]["content"], "\n")
```

#### Text moderation

* Text moderation is the process of identifying text that is inappropriate for the context it is being used in. In online communities like social networks and chat rooms, moderation is crucial to prevent the exchange of harmful and offensive content. 
    * word detection algorithm -- does not understand context

```python
response = openai.Moderation.create(
    model = "text-moderation-latest",
    input = "Gerçekten bu işi alabilmek için adam öldürmemi falan istiyorlar sanki.."
)
print(response)

##category_scores are float values for each category indicating the model's confidence of a violation (0-1 but not probability)
```

```python
response["results"][0]["category_scores"] # determining thresholds
```

```python
## adding in a prompt
audio_file = open("non-english-audio.mp3", "rb") ## rb: read binary
prompt = "The transcript is about AI trends and ChatGPT"
response = openai.Audio.translate("whisper-1", 
audio_file,
prompt = prompt
)
print(response["text"])
```

```python
## Combining models
# chaining: feeding output from one model into another

audio_file = open("english-audio.mp3", "rb") ## rb: read binary
audio_response = openai.Audio.transcribe("whisper-1", audio_file)
transcript = audio_response["text"]
prompt = "Extract the attendee names from the start of this meeting transcript: " + transcript
chat_response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [{
        "role": "user", "content":prompt
    }]
)
print(chat_response["choices"][0]["message"]["content"])
```
